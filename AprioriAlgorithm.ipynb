```python
#***César Romero Canul_ DataE***

#Calling library "collections"
from collections import Counter           

#Defining Apriori function 
def apriori(Transactions, minSupport):      
    #For each element in Transactions array, items will be counted.
    Counter = {}
    
    for Groups in Transactions:
        for Items in Groups:
            if Items in Counter:
                Counter[Items] += 1
            else:
                Counter[Items] = 1
    #Print the frequency of elements in Transactions.
    print ("Items_Frequency: \n", Counter)
    
    #FI meas Frequent Items; Counter.keys(),in this case, prints just the element without frequency.
    FI = Counter.keys()
    items = []
    #Putting all the elements in a new array
    for x in FI:
        items.append([x])
    #According to the support given, define the Frequent Items, putting all them into an a new array.
    FreqItems = []
    for y in items[:]:
        if Counter[y[0]] >= minSupport:
            FreqItems.append(y)
    FreqItems.sort()
    print ("\nFrequent_Items:\n", FreqItems, "\n")
    
    Results = [] 
    #Parameters
    i = 1 
    j = 2

    #Loop used to gem¿nerate the final results
    while FreqItems != []:
        Freq = Frequency(Transactions, FreqItems)          #Calling Frequency function
        FreqTrans = Remove(FreqItems, Freq, minSupport)    #Calling Remove function
        
        for x in FreqTrans:
            Results.append(x)                              
        print("\nFrequent groups of ",i," elements: ",FreqTrans)            #Printing  frequent groups of transactions
        print("\n")
        i = i+1
        print("---------------------------------------------------------")
        print("Groups of {}:".format(j))                                    #Combining items 
        j = j+1
        FreqItems = Combine(FreqTrans, FreqItems)                      #Calling Combine function
    return Results


#Defining Frequency function
def Frequency(Transactions, FreqItems):
    Freq = []
    #Verifying the times elements in FreqItmes appear
    for x in FreqItems:
        c = 0
        for y in Transactions:
            it = True
            for z in x:
                if z not in y:
                    it = False
            if it:
                c = c+1
        Freq.append(c) #Adding the number they appear into Freq array
    return Freq


#Defining Remove function
def Remove(FreqItems, Freq, minSupport):
    newFI= []
    #Printing frequency of each Frequent Item
    for q, val in enumerate(FreqItems):
        print("For {} frequency is: {}".format(val,Freq[q]))
        #Removing elements which their frequency is lower than the given support
        if Freq[q] >= minSupport:
            newFI.append(FreqItems[q])
    return newFI


#Defining Combine function
def Combine(FreqTrans, FreqItems):
    FreqTrans.sort()
    Groups = []
    #Combining process
    for x in FreqTrans:
        for y in FreqTrans:
            if (x != y):
                sets = []
                t = len(x)
                #Adding elements to sets
                if (t) == 1:
                    for val in x:
                        if (val not in sets):
                            sets.append(val)
                    for val in y:
                        if (val not in sets):
                            sets.append(val)
                            sets.sort()
                    #Adding sets to Groups array
                    if (sets not in Groups):
                        Groups.append(sets)
                        Groups.sort()
                            
                #Adding elements to sets when combinations must have more than 2 of them           
                elif (t) > 1 :
                    for a in x:
                        for b in y:
                            if a == b:
                                for val in x:
                                    if (val not in sets):
                                        sets.append(val)
                                for val in y:
                                    if (val not in sets):
                                        sets.append(val)
                                        sets.sort()
                                #Adding sets to Groups array       
                                if (sets not in Groups):
                                    Groups.append(sets)
                                    Groups.sort()
       
                            elif a != b:
                                continue   
    print(Groups)
    print("\n")
    return Groups


#Executing main program
if __name__ == "__main__":
    
    Transactions = [['water','honey','milk','coke','salt'],
                    ['oil','popcorn','water','salt','milk','sabritas'],
                    ['sugar','oil','milk','water','ketchup'],
                    ['nuts','water','onions','salt','carrots','milk'],
                    ['salt','coke','milk','chocolate','honey','popcorn'],
                    ['carrots','onions','honey','milk','chocolate'],
                    ['potatoes','carrots','water','onions','milk','salt'],
                    ['honey','oil','coke','salt','milk','water']]
    
    print("///////The Apriori Algorithm///////\n")
    minSupport = int(input("Support: "))
    print("\n")

    #Calling Apriori function
    FreqTrans = apriori(Transactions, minSupport)
```
